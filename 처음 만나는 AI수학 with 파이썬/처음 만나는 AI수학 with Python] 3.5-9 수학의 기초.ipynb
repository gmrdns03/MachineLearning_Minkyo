# 3.5 삼각함수
삼각함수를 이용하면 주기적이며 매끄럽게 변화하는 값을 다룰 수 있다.
sinθ, sosθ, tanθ를 삼각함수라고 한다.

각도 θ의 단위는 라디안을 자주 사용한다.

## 352_ 삼각함수를 구현
y = sinX, y= cosX를 코드로 구현하기<br>
numpy의 sin(), cos()함수를 이용<br>
인수의 단위는 라디안<br>
원주율은 np.pi로 얻을 수 있다.<br>
sin()함수는 cos()함수를 x방향으로 pi/2만큼 시프트된 것

```python
# 리스트 3.15 sin() 함수와 cos() 함수를 그래프로 그린다.

import numpy as np
import matplotlib.pyplot as plt
# % matplotlib inline

def my_sin(x):
    return np.sin(x) # sin(x)

def my_cos(x):
    return np.cos(x) # cos(x)

x = np.linspace(-np.pi, np.pi)  
y_sin = my_sin(x)
y_cos = my_cos(x)

plt.plot(x, y_sin, label="sin")
plt.plot(x, y_cos, label="cos")
plt.legend()

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()

plt.show()
```

```python
# 리스트 3.13 tan()함수를 그래프로 그린다.

import numpy as np
import matplotlib.pyplot as plt

def my_tan(x):
    return np.tan(x) # tan(x)

x = np.linspace(-1.3, 1.3) # -1.3부터 1.3라디안 까지 
y_tan = my_tan(x)

plt.plot(x,y_tan, label="tan")
plt.legend()

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()

plt.show()
```

# 3.6 총합과 총곱

## 361_ 총합

총합은 여러 개의 수치를 모두 더하는 걸 말한다.

## 362_ 총합을 구현
총합은 numpy의 sum()함수를 사용해서 간단히 구할 수 있다.

```python
# 리스트 3.19 Numpy의 sum() 함수의 총합을 구한다.

import numpy as np

a = np.array([1,3,2,5,4])  # a1부터 a5까지
y = np.sum(a)  # 총합
print(y)
```

## 363_ 총곱
총 곱은 여러개의 수치를 모두 곱한 것을 말한다.

## 364_ 총곱을 구현
총곱은 numpy의 prod()함수를 사용해서 구할 수 있다.

```python
# 리스트 3.20 NumPy의 prod()함수로 총 곱을 구한다.

import numpy as np

a = np.array([1,3,2,5,4])  # a1부터 a5까지
y = np.prod(a) # 총곱
print(y)
```

# 3.7 난수
난수는 규칙성이 없는 예측할 수 없는 수치이다.

인공지능에서 파리미터의 초기화 등에 난수가 활용된다.

## 371_ 난수

```python
# 리스트 3.23 1부터 6까지 정수의 난수를 생성한다.
import numpy as np

r_int = np.random.randint(6) + 1  # 0부터 5까지의 난수에 1을 더한다.
print(r_int) # 1부터 6까지가 랜덤으로 표시된다.
```

```python
# 리스트 3.24 0부터 1까지의 소수의 난수를 생성한다.
import numpy as np

r_dec = np.random.rand()  # 0부터 1사이의 소수를 랜덤으로 반환한다.
                           # ()안에 숫자를 써주면 숫자 만큼 리스트로 소수를 반환
print(r_dec)
```

## 372_ 균일한 난수
앞의 random.rand() 함수는 0에서 1사이의 소수를 균등한 확률로 반환
()안에 정수 a를 인수로 건네면 a개의 난수가 균일한 확률로 반환

```python
# 리스트 3.25 난수의 균일한 분포
import numpy as np
import matplotlib.pyplot as plt

n = 1000  # 샘플 수
x = np.random.rand(n) # 0~1 사이의 균일한 난수
y = np.random.rand(n) # 0~1 사이의 균일한 난수

plt.scatter(x, y)     # 산포도를 플롯
plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()

plt.show()

```

## 373_ 편향된 난수
난수가 결정될 확률은 균일하다고는 할 수 없다.

random.randn()함수는 정규 분포라는 분포를 따른느 확률로 난수를 반환한다.

```python
# 리스트 3.26 정규분포를 따르는 난수의 분포
import numpy as np
import matplotlib.pyplot as plt

n = 1000  # 샘플 수
x = np.random.randn(n) # 정규분포를 따른 난수의 분포
y = np.random.randn(n) # 정규분포를 따른 난수의 분포

plt.scatter(x, y)     # 산포도를 플롯
plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()

plt.show()
```

# 3.8 LaTeX의 기초
라텍이라는 문서 처리 시스템을 사용해 수식을 깨끗하게 입력하는 방법을 배운다.

## 381_ LaTeX
주피터 노트북은 라텍이라는 문서 처리 시스템을 이용해서 수식을 표현할 수 있다.

### 리스트 3.29 기술 예
$$y = 2x+1$$

## 382_ 여러가지 수식의 기술
- 첨자와 거듭제곱<br>
a_1 -> $$a_1$$ <br>
a_{ij} -> $$a_{ij}$$ <br>
b^2 -> $$b^2$$ <br>
b^{ij} -> $$b^{ij}$$ <br>
c_1^2 -> $$c_1^2$$ <br>
<br>
- 다항식 <br>
y=x^3+2x^2+x+3 -> $$y=x^3+2x^2+x+3$$ <br>
<br>
- 제곱근 <br>
\sqrt를 이용해 루트 기호를 기술할 수 있다. <br>
y = \sqrt x -> $$y = \sqrt x$$  <br>
<br>
- 삼각함수 <br>
\sin 이나 \cos을 이용해 삼각함수를 입력할 수 있다. <br>
y=\sin x -> $$y=\sin x$$ <br>
y=\cos x -> $$y=\cos x$$ <br>
y=\tan x -> $$y=\tan x$$ <br>
<br>
- 분수
\frac{}{}을 이용해 분수를 기술할 수 있다. <br>
y=/frac{17}{24} -> $$y=/frac{17}{24}$$ <br>
<br>
- 총합<br>
\sum을 이용해 시그마 기호를 기술할 수 있다.<br>
y=\sum_{k=1}^n a_k -> $$y=\sum_{k=1}^n a_k$$<br>
<br>
- 총곱<br>
\prod을 이용해 총곱 기호를 기술할 수 있다.<br>
y=\prod_{k=1}^n a_k -> $$y=\prod_{k=1}^n a_k$$<br>

# 3.9 절댓값
절댓값은 값과 0사이의 거리를 나타낸다.

인공지능에서 0을 중심으로 한 값의 퍼짐 상태를 파악하기 위해서 사용되는 경우가 있다.

## 391_ 절댓값
절댓값은 numpy의 abs() 함수를 사용해 구할 수 있다.

```python
# 리스트 3.31 abs()함수를 사용해 절댓값을 구한다.
import numpy as np
x = [-5, 5, -1.28, np.sqrt(5), -np.pi/2]
print(np.abs(x))  # 절댓값을 구한다.
```

## 391_ 함수의 절댓값

```python
# 리스트 3.32 삼각함수의 절댓값
import numpy as np
import matplotlib.pyplot as plt
# % matplotlib inline

def my_sin(x):
    return np.sin(x) # sin(x)

def my_cos(x):
    return np.cos(x) # cos(x)

x = np.linspace(-np.pi, np.pi)  
y_sin = np.abs(my_sin(x))
y_cos = np.abs(my_cos(x))

plt.plot(x, y_sin, label="sin")
plt.plot(x, y_cos, label="cos")
plt.legend()

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()

plt.show()
```
