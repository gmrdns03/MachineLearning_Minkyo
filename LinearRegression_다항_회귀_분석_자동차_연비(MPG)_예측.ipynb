LinearRegression 다항 회귀 분석을 사용 하여 1970년대 후반과 1980년대 초반의 자동차 연비를 예측

```python
# 기본 라이브러리 불러오기
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## [step 1] 데이터 준비
- read_csv() 함수로 자동차 연비 데이터셋 가져오기

```python
# csv 파일을 데이터프레임으로 변환
df = pd.read_csv('/content/auto-mpg.csv', header=None)
```

```python
# 열 이름 지정
df.columns = ['mpg','cylinders','displacement','horsepower','weight',
              'acceleration','model year','origin','name']
```

```python
df.head()
```

## [step 2] 데이터 탐색 및 데이터 전처리

```python
df.info()
```

```python
# horsepower 열의 자료형 변경(문자열 -> 숫자)
df['horsepower'].unique()
```

```python
# '?'로 되어 있는 오브젝트를 np.nan으로
df['horsepower'].replace('?', np.nan, inplace=True)
df.info()
```

```python
# 누락 데이터를 삭제
df.dropna(subset=['horsepower'], axis=0, inplace=True)
df.info()
```

```python
# 문자열 -> 숫자로 변환
df['horsepower'] = df['horsepower'].astype('float')
df.info()
```

## [step 3] 속성(feature 또는 variable)선택

```python
# 분석에 활용할 속성 선택(연비, 실린더, 마력, 중량)
ndf = df[['mpg', 'cylinders', 'horsepower', 'weight']]
ndf.head()
```

## [step 4] 데이터셋 구분

- 훈련용(train data)/검증용(test data)

```python
# 속성(변수) 선택 -> 단순회귀
# 독립변수 x1, x2, x3
X = ndf[['cylinders','horsepower', 'weight']]  # X는 2차원으로 만들어줘야 하기 때문에 [[]]라고 썻음
y = ndf['mpg']  # 종속변수 y
```

```python
# seaborn pairplot으로 두 변수간의 모든 경우의 수 그리기
sns.pairplot(ndf)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,               #  독립변수
    y,               #  종속변수
    test_size=0.3,   #  검증 30%
    random_state=10) #  랜덤 추출 값
```

## [step 5] 다항 회귀 분석 모형
- sklearn 사용

```python
# sklearn 라이브러리에서 선형회귀분석 모듈 가져오기
from sklearn.linear_model import LinearRegression

# 단순 회귀분석 모형객체 생성 후 학습시키기
lr = LinearRegression().fit(X_train, y_train)
```

## [step 6] 성능 평가하기

```python
# 학습을 마친 모형에 test data를 적용하여 결정계수(R-제곱)계산 -> 결정계수 값이클수록 모형의 예측 능력이 좋다고 판단.
r_square = lr.score(X_test, y_test)
r_square
```

```python
# 모형에 전체 X데이터를 입력하여 예측한 값 y_hat으로 담기
y_hat = lr.predict(X)
```

```python
# 모형에 테스트셋 X_test 데이터를 입력하여 예측한 값 y_hat_test으로 담기
y_hat_test = lr.predict(X_test)
```

```python
# train 데이터(X_train)의 산점도와 test 데이터로 예측한(y_hat_test) 회귀선을 그래프로 출력
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(1,1,1)

ax.plot(X_train, y_train, 'o', label='Train Data')  # 데이터의 분포
ax.plot(X_test, y_hat_test, 'r+', label='Predicted Value')  # 모형이 예측한 회귀선

ax.legend(loc='best')  # best라고 하면 빈 공간을 찾아서 알아서 넣어준다.

plt.xlabel('features')
plt.ylabel('mpg')

plt.show()

# 테스트 결과가 실재를 어느정도는 이해했다는 것을 전체적인 양상을 통해 확인해보기
```

```python
print("훈련세트 정확도: {:.2f}%".format(lr.score(X_train, y_train)*100))
print("테스트세트 정확도: {:.2f}%".format(lr.score(X_test, y_test)*100))
```

```python
plt.figure(figsize= (10, 5))
ax1 = sns.kdeplot(y, label='y')
ax2 = sns.kdeplot(y_hat, label='y_hat', ax=ax1)  # ax=ax1 ax1에 같이 그려달라는 뜻
plt.legend()
plt.title("LinearRegression 다항 회귀 분석 자동차 연비(MPG)예측")

plt.show()
```
